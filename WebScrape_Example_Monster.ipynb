# Example Web Scrape Monster

```python
#import the package needed to pull the html information
import requests
#import pandas for everything else
import pandas as pd
#import Beautiful soup to help make sense of the html tags
from bs4 import BeautifulSoup
```

```python
#Retrieve the HTMl data from the site and store in an object
URL = 'https://www.monster.com/jobs/search/?q=Software-Developer&where=Australia'
page = requests.get(URL)
#Apply Beautiful soup html parser 
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
#Pull the specific element you need
results = soup.find(id='ResultsContainer')
```

```python
#prettify cleans up the html code
print(results.prettify())
```

```python
# Leveage <section> element with the class card-content. to select only the job listings
job_elems = results.find_all('section', class_='card-content')
```

```python
#let's take a look at all of them
#Notice still lots of html present
for job_elem in job_elems:
    print(job_elem, end='\n'*2)
```

```python
#Now lets pull the descriptive class names
for job_elem in job_elems:
    title_elem = job_elem.find('h2', class_='title')
    company_elem = job_elem.find('div', class_='company')
    location_elem = job_elem.find('div', class_ = 'location')
    print(title_elem)
    print(company_elem)
    print(location_elem)
    print()
```

```python
#We want to see just the title, company and location of each job posting. Beautiful soup can do that
#Adding .text to a beautiful soup object can return only the text content of the HTML elements that the object contains

for job_elem in job_elems:
    title_elem = job_elem.find('h2', class_='title')
    company_elem = job_elem.find('div', class_='company')
    location_elem = job_elem.find('div', class_ = 'location')
    print(title_elem.text)
    print(company_elem.text)
    print(location_elem.text)
    print()
```

```python
#added control for None
#used .strip to clear the whitespaces
for job_elem in job_elems:
    title_elem = job_elem.find('h2', class_='title')
    company_elem = job_elem.find('div', class_='company')
    location_elem = job_elem.find('div', class_='location')
    if None in (title_elem, company_elem, location_elem):
        continue
    print(title_elem.text.strip())
    print(company_elem.text.strip())
    print(location_elem.text.strip())
    print()
```

```python
#Now lets turn this into a dataframe so we can use it for analysis
d =[]
for job_elem in job_elems:
    title_elem = job_elem.find('h2', class_='title')
    company_elem = job_elem.find('div', class_='company')
    location_elem = job_elem.find('div', class_='location')
    if None in (title_elem, company_elem, location_elem):
        continue
    title =title_elem.text.strip()
    Company =company_elem.text.strip()
    Location =location_elem.text.strip()
    d.append({'Job_Title' :title, 'Company' :Company, 'Location' :Location})
pd.DataFrame(d)
```
